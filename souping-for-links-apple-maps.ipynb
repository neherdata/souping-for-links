# Souping for Links [All]


```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://gspe21-ssl.ls.apple.com/html/attribution-259.html'
r = requests.get(url)
r.status_code
```

```python
soup = BeautifulSoup(r.text)

links = []

for link in soup.find_all('a'):
    current_link = link.get('href')
    links.append(current_link)

len(links)
```
